# yawn-dataset

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, LSTM, TimeDistributed
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

# Set random seed for reproducibility
seed_value = 42
np.random.seed(seed_value)
tf.random.set_seed(seed_value)

# Load and preprocess images (32x32 resolution)
def load_images(path, target_shape=(32, 32)):
    images = []
    for filename in os.listdir(path):
        if filename.endswith(".jpg") or filename.endswith(".png"): 
            img = tf.keras.preprocessing.image.load_img(os.path.join(path, filename), target_size=target_shape)
            img = tf.keras.preprocessing.image.img_to_array(img)
            images.append(img)
    return np.array(images)

# Example paths (adjust to your dataset paths)
yawn_path = '/kaggle/input/yawn-dataset/yawn'
no_yawn_path = '/kaggle/input/yawn-dataset/no yawn'

# Load data
yawn_images = load_images(yawn_path)
no_yawn_images = load_images(no_yawn_path)

# Combine data and create labels
X = np.concatenate([yawn_images, no_yawn_images], axis=0)
y_labels = np.array([0] * len(yawn_images) + [1] * len(no_yawn_images))

# Normalize image data to [0, 1] range
X = X.astype('float32') / 255.0

# One-hot encode the labels
y_labels = to_categorical(y_labels)

# Reshape data into a sequence format suitable for LSTM
# Assuming each image is a single frame in a sequence, reshape to (samples, time steps, width, height, channels)
X = X.reshape(-1, 1, 32, 32, 3)  # 1 time step per image

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_labels, test_size=0.2, random_state=seed_value)

# CNN-LSTM Model Definition
model = Sequential()

# TimeDistributed wrapper to apply CNN on each time step
model.add(TimeDistributed(Conv2D(32, (3, 3), activation='relu', padding='same'), input_shape=(1, 32, 32, 3)))
model.add(TimeDistributed(MaxPooling2D(pool_size=(2, 2))))
model.add(TimeDistributed(Dropout(0.3)))

model.add(TimeDistributed(Conv2D(64, (3, 3), activation='relu', padding='same')))
model.add(TimeDistributed(MaxPooling2D(pool_size=(2, 2))))
model.add(TimeDistributed(Dropout(0.3)))

model.add(TimeDistributed(Flatten()))

# LSTM Layer to learn temporal dependencies
model.add(LSTM(64, activation='relu', return_sequences=False))
model.add(Dropout(0.5))

# Dense layer for classification
model.add(Dense(64, activation='relu'))
model.add(Dense(2, activation='softmax'))  # 2 output classes (yawn or no_yawn)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Plot training history (loss and accuracy)
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    epochs_range = range(len(acc))

    plt.figure(figsize=(14, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.show()

# Plot the training history
plot_training_history(history)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_acc}")

# Save the model if needed
model.save('cnn_lstm_model.h5')
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, LSTM, TimeDistributed
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

# Set random seed for reproducibility
seed_value = 42
np.random.seed(seed_value)
tf.random.set_seed(seed_value)

# Load and preprocess images (32x32 resolution)
def load_images(path, target_shape=(32, 32)):
    images = []
    for filename in os.listdir(path):
        if filename.endswith(".jpg") or filename.endswith(".png"): 
            img = tf.keras.preprocessing.image.load_img(os.path.join(path, filename), target_size=target_shape)
            img = tf.keras.preprocessing.image.img_to_array(img)
            images.append(img)
    return np.array(images)

# Load datasets from three different paths
yawn_path = '/kaggle/input/yawn-eye-dataset-new/dataset_new/train/yawn'
no_yawn_path = '/kaggle/input/yawn-eye-dataset-new/dataset_new/train/no_yawn'
open_images_path = '/kaggle/input/yawn-eye-dataset-new/dataset_new/train/Open'
close_images_path = '/kaggle/input/yawn-eye-dataset-new/dataset_new/train/Closed'
down_path = '/kaggle/input/face-pose-dataset/down'
left_path = '/kaggle/input/face-pose-dataset/left'
right_path = '/kaggle/input/face-pose-dataset/right'
straight_path = '/kaggle/input/face-pose-dataset/straight'
up_path = '/kaggle/input/face-pose-dataset/up'

# Load data
yawn_images = load_images(yawn_path)
no_yawn_images = load_images(no_yawn_path)
open_images = load_images(open_images_path)
close_images = load_images(close_images_path)
down_images = load_images(down_path)
left_images = load_images(left_path)
right_images = load_images(right_path)
straight_images = load_images(straight_path)
up_images = load_images(up_path)

# Combine all data and create labels
X = np.concatenate([
    yawn_images, no_yawn_images, open_images, close_images,
    down_images, left_images, right_images, straight_images, up_images
], axis=0)

# Create labels for all classes
y_labels = np.array([0] * len(yawn_images) +     # Yawn
                    [1] * len(no_yawn_images) + # No Yawn
                    [2] * len(open_images) +    # Open
                    [3] * len(close_images) +   # Closed
                    [4] * len(down_images) +    # Down
                    [5] * len(left_images) +    # Left
                    [6] * len(right_images) +   # Right
                    [7] * len(straight_images) + # Straight
                    [8] * len(up_images)        # Up
                   )

# Normalize image data to [0, 1] range
X = X.astype('float32') / 255.0

# One-hot encode the labels (for 9 classes)
y_labels = to_categorical(y_labels, num_classes=9)

# Reshape data into a sequence format suitable for LSTM
X = X.reshape(-1, 1, 32, 32, 3)  # 1 time step per image

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_labels, test_size=0.2, random_state=seed_value)

# Check shapes after the split
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")

# CNN-LSTM Model Definition
model = Sequential()

# TimeDistributed wrapper to apply CNN on each time step
model.add(TimeDistributed(Conv2D(32, (3, 3), activation='relu', padding='same'), input_shape=(1, 32, 32, 3)))
model.add(TimeDistributed(MaxPooling2D(pool_size=(2, 2))))
model.add(TimeDistributed(Dropout(0.3)))

model.add(TimeDistributed(Conv2D(64, (3, 3), activation='relu', padding='same')))
model.add(TimeDistributed(MaxPooling2D(pool_size=(2, 2))))
model.add(TimeDistributed(Dropout(0.3)))

model.add(TimeDistributed(Flatten()))

# LSTM Layer to learn temporal dependencies
model.add(LSTM(64, activation='relu', return_sequences=False))
model.add(Dropout(0.5))

# Dense layer for classification
model.add(Dense(64, activation='relu'))

# Output layer: 9 units for 9 classes (yawn, no yawn, open, closed, down, left, right, straight, up)
model.add(Dense(9, activation='softmax'))  # 9 output classes

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Plot training history (loss and accuracy)
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    epochs_range = range(len(acc))

    plt.figure(figsize=(14, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.show()

# Plot the training history
plot_training_history(history)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_acc}")

# Save the model if needed
model.save('cnn_lstm_combined_model.h5')
```

```python

# Function to preprocess the input image for prediction
def preprocess_input_image(img_path, target_size=(32, 32)):
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=target_size)  # Load the image
    img_array = tf.keras.preprocessing.image.img_to_array(img)  # Convert image to array
    img_array = img_array.astype('float32') / 255.0  # Normalize image data to [0, 1]
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension (1, 32, 32, 3)
    img_array = np.reshape(img_array, (1, 1, 32, 32, 3))  # Reshape for CNN-LSTM input (1, 1, 32, 32, 3)
    return img_array

# Function to display the image and prediction result
def display_prediction(img_path, result):
    img = tf.keras.preprocessing.image.load_img(img_path)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Predicted: {result}")
    plt.show()

# Make a prediction
def predict_image_class(img_path):
    input_image = preprocess_input_image(img_path)
    prediction = model.predict(input_image)
    predicted_class = np.argmax(prediction)
    
    # Map the prediction to class labels
    class_labels = {
        0: 'Yawn',
        1: 'No Yawn',
        2: 'Open',
        3: 'Closed',
        4: 'Down',
        5: 'Left',
        6: 'Right',
        7: 'Straight',
        8: 'Up'
    }
    result = class_labels[predicted_class]
    
    print(f"Predicted Class: {result}")
    display_prediction(img_path, result)

# Example of using the prediction function
# Replace the path with an actual image path to test the prediction
img_path = '/kaggle/input/face-pose-dataset/left/Angela_Lansbury_0002.jpg'  # Update with your test image path
predict_image_class(img_path)
```

```python
# Function to make predictions and return all matching classes
def predict_all_classes(img_path, threshold=0.5):
    input_image = preprocess_input_image(img_path)
    prediction = model.predict(input_image)
    
    # Get all class probabilities
    class_probabilities = prediction[0]  # Since the output is (1, num_classes), take the first element
    
    # Prepare the class labels
    class_labels = {
        0: 'Yawn',
        1: 'No Yawn',
        2: 'Open',
        3: 'Closed',
        4: 'Down',
        5: 'Left',
        6: 'Right',
        7: 'Straight',
        8: 'Up'
    }
    
    # Find classes above the confidence threshold
    matching_classes = {class_labels[i]: prob for i, prob in enumerate(class_probabilities) if prob >= threshold}
    
    if matching_classes:
        print("Predicted Classes and Probabilities:")
        for cls, prob in matching_classes.items():
            print(f"{cls}: {prob:.2f}")
    else:
        print("No classes meet the confidence threshold.")

    display_prediction(img_path, matching_classes)

# Example of using the prediction function
img_path = '/kaggle/input/yawn-eye-dataset-new/dataset_new/train/no_yawn/1021.jpg'  # Update with your test image path
predict_all_classes(img_path, threshold=0.5)  # Adjust the threshold as needed
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to make predictions and display a bar graph of all class probabilities
def predict_and_plot_classes(img_path, threshold=0.5):
    input_image = preprocess_input_image(img_path)
    prediction = model.predict(input_image)
    
    # Get all class probabilities
    class_probabilities = prediction[0]  # Since the output is (1, num_classes), take the first element
    
    # Prepare the class labels
    class_labels = {
        0: 'Yawn',
        1: 'No Yawn',
        2: 'Open',
        3: 'Closed',
        4: 'Down',
        5: 'Left',
        6: 'Right',
        7: 'Straight',
        8: 'Up'
    }
    
    # Filter classes based on the threshold
    matching_classes = {class_labels[i]: prob for i, prob in enumerate(class_probabilities) if prob >= threshold}
    
    if matching_classes:
        print("Predicted Classes and Probabilities:")
        for cls, prob in matching_classes.items():
            print(f"{cls}: {prob:.2f}")
    else:
        print("No classes meet the confidence threshold.")
    
    # Plotting the class probabilities
    plt.figure(figsize=(10, 6))
    plt.bar(class_labels.values(), class_probabilities, color='skyblue')
    plt.xlabel('Classes')
    plt.ylabel('Probability')
    plt.title('Class Probability Distribution for Input Image')
    plt.ylim(0, 1)  # Probability range from 0 to 1
    plt.xticks(rotation=45)
    plt.axhline(y=threshold, color='r', linestyle='--', label='Threshold')
    plt.legend()
    plt.show()
    
    display_prediction(img_path, matching_classes)

# Example of using the prediction function with visualization
img_path = '/kaggle/input/yawn-eye-dataset-new/dataset_new/train/no_yawn/1021.jpg'  # Update with your test image path
predict_and_plot_classes(img_path, threshold=0.5)  # Adjust the threshold as needed
```
